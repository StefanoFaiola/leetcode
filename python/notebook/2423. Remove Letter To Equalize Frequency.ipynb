# 2423. Remove Letter To Equalize Frequency

```python
word = 'aazzz'
```

```python
def equalFrequency(word):
    word_map = {}
    
    for k in word:
        word_map[k] = word.count(k)

    for i in word_map:
        word_list = []
        word_map[i] -= 1
        
        for _ in word_map.values():
            if _ > 0:
                word_list.append(_)

        if len(set(word_list)) == 1:
            return True
        else:
            word_map[i] += 1
    return False

    

    














```

```python
equalFrequency(word)
```
